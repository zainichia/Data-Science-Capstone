```python
import pandas as pd
import requests
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
df[0]
```

```python
df_postalcodes = df[0]
```

```python
df_postalcodes.head()
```

```python
df_postalcodes["Borough"].value_counts()
```

```python
cond_borough = (df_postalcodes["Borough"] != "Not assigned")
```

```python
df_postalcodes = df_postalcodes[cond_borough]
```

```python
df_postalcodes.shape
```

```python
df_postalcodes.head(10)
```

```python
df_postalcodes.reset_index(drop = True, inplace = True)
```

```python
df_postalcodes.head(10)
```

```python
cond_nbhd = df_postalcodes["Neighbourhood"] == "Not assigned"
```

```python
cond_nbhd.value_counts()
```

```python
df_postalcodes.iloc[6, 2] = df_postalcodes.iloc[6, 1]
```

```python
df_postalcodes.head(10)
```

```python
postal_codes = df_postalcodes["Postcode"].value_counts().to_frame()
```

```python
postal_codes.shape
```

```python
postal_codes.head()
```

```python
df_postalcodes_grouped = df_postalcodes.groupby(["Postcode", "Borough"])["Neighbourhood"].apply(', '.join)
```

```python
df_postalcodes_grouped = df_postalcodes_grouped.to_frame()
```

```python
df_postalcodes_grouped.reset_index(inplace = True)
```

```python
df_postalcodes_grouped.head(10)
```

```python
df_postalcodes_grouped.shape
```

```python
df_postalcodes_grouped.sort_values(by = "Postcode")
```

```python
df_postalcodes_grouped.head()
```

```python
df_postalcodes_grouped.shape
```

```python
df_LocationData = pd.read_csv("https://cocl.us/Geospatial_data")
```

```python
df_LocationData.head()
```

```python
cond = df_LocationData["Postal Code"] == df_postalcodes_grouped["Postcode"]
```

```python
cond.value_counts()
```

```python
df_postalcodes_final = pd.concat([df_postalcodes_grouped, df_LocationData], axis = 1, join = 'inner')
```

```python
df_postalcodes_final.head()
```

```python
df_postalcodes_final.drop(["Postal Code"], axis = 1, inplace = True)
```
